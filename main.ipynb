## Initiation

```python
import pandas as pd
from tqdm import tqdm
import os
import spotify
import genius

# Make sure there is a data directory
if not os.path.exists('./data'):
    os.makedirs('./data')
```

## Spotify

### requesting top 1000 popular songs on 2021

```python
df_track = spotify.require_popular_songs(2021,1000)
df_track
```

### requesting and assign artist informations to the original dataframe

```python
df_track = spotify.assign_artist_info(df_track)
df_track
```

### requesting and assign track features to the original dataframe

```python
df_track = spotify.assign_track_features(df_track)
df_track
```

### change on data type

```python
df_track['artist_name'] = df_track['artist_name'].astype("string")
df_track['track_name'] = df_track['track_name'].astype("string")
df_track['track_id'] = df_track['track_id'].astype("string")
df_track['artist_id'] = df_track['artist_id'].astype("string")
df_track['duration_ms'] = pd.to_numeric(df_track['duration_ms'])
df_track['instrumentalness'] = pd.to_numeric(df_track['instrumentalness'])
```

### The final output of the popular track informations

```python
df_track.info()
```

```python
# write the dataframe to a csv file
df_track.to_csv('data/track.csv',index = False,encoding = 'UTF-8')
```

## Genius

```python
# read back the track df if not loaded
df_track = pd.read_csv('data/track.csv')
```

```python
# initiate the lyrics data frame
df_lyrics = df_track[['artist_name','track_name','track_id','artist_id']]
# write the lyrics to the dataframe
def apply_get_lyrics(x):
    return genius.get_lyrics(x['artist_name'],x['track_name'])
df_lyrics['lyrics'] = df_lyrics.apply(apply_get_lyrics,axis=1)
```

```python
df_lyrics.info()
```

```python
# write the lyrics to a csv file
df_lyrics.to_csv('data/lyrics.csv',index = False,encoding = 'UTF-8')
```
